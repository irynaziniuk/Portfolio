## Завдання 2. Аналіз повноти відповідей респондентів

```python
questions = set(df1['qname']) # список питань df1

df_columns = set(df.columns)
common_questions = questions.intersection(df_columns) # спільні питання df і df1
respondents_full_answers = df.dropna(subset=common_questions) # відповіді на всі питання
count_full_respondents = respondents_full_answers.shape[0] # к-ть респондентів, що відповіли на всі запитання
print (f'{count_full_respondents} респондентів відповіли на всі запитання')
```

 ## Завдання 3. Статистичний аналіз досвіду респондентів

```python
mean_workexp = round(df['WorkExp'].mean(), 1) # Середнє значення (mean)
median_workexp = df['WorkExp'].median() # Медіана (median)
mode_workexp = df['WorkExp'].mode()[0] # Мода (mode)
print (f'Міри центральної тенденції : Середнє значення = {mean_workexp}, Медіана = {median_workexp}, Мода = {mode_workexp}')
# Висновок: 
# Медіана = 10 р досвіду, що свідчить про те, що половина респондентів має досвід роботи менше або дорівнює 10 років.
# Мода = 10 р досвіду, що свідчить про те, найбільше респондентів має досвід роботи 10 років.
# Середнє значення = 13,4 р досвіду. Сережнє значення більше за медіану, це вказує на респонедентів з великим досвідом, які тягнуть середнє вгору.
```

 ## Завдання 5. Визначення популярності Python

 ## Завдання 7. Географічний аналіз компенсації Python-розробників

```python
worked_with_python = df[df['LanguageHaveWorkedWith'].str.contains('Python', na = False)]
worked_with_python = worked_with_python.dropna(subset = ['ConvertedCompYearly', 'Country']) # видалення відсутніх значень комепенсації
grouped_by_country = worked_with_python.groupby('Country') # групування за країнами
mean_compensation = round(grouped_by_country['ConvertedCompYearly'].mean(), 2) # середнє
median_compensation = round(grouped_by_country['ConvertedCompYearly'].median(), 2) # медіана
py_progres_by_country = pd.DataFrame({
    'mean_compensation': mean_compensation,
    'median_compensation': median_compensation
})
py_progres_by_country
```

## *Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
df['Python'] = df['LanguageHaveWorkedWith'].str.contains('Python', na = False)
df_age = df.dropna(subset = ['Age']) # видалення відсутніх значень віку для респондентів
python_count = df_age[df_age['Python'] == True].groupby('Age')['Python'].sum() # к-ть респондентів, що програмують на python у кожній віковій групі, 
total_count = df_age.groupby('Age')['Python'].count() # загальна к-ть респондентів у кожній віковій групі
python_percent_group_age = round((python_count / total_count) * 100, 2) # % Python-розробників по віковим групам
python_percent_group_age = pd.DataFrame({'% Python-розробників': python_percent_group_age})
python_percent_group_age

# Висновок: Найбільше Python-розробників у віковій групі 18-24 роки, а найменше - 65 і більше років.
```

## *Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
Q3 = df['ConvertedCompYearly'].quantile(0.75) #75-й перцентиль:
high_paid_remote = df[
    (df['ConvertedCompYearly'] >= Q3) &
    (df['RemoteWork'] == 'Remote')] # Респонденти, які заробляють більше, ніж 75% респондентів і працюють віддалено
industry_counts = high_paid_remote['Industry'].value_counts()
industry_counts

# Висновок: Найрозповсюдженіші індустрі: Software Development - 1186, Fintech - 190, Healthcare - 188.
```
